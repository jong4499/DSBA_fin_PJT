### 요약문 1차 맞춤법 교정

```python
check_lst = []

for text in testdf:
    hanspell_sent = spell_checker.check(text)
    # 리스트에 교정한 요약문 추가
    check_lst.append(hanspell_sent.checked)
```

```python
# KoELECTRA 모델 및 토크나이저 로드
model_name = "monologg/koelectra-base-v3-discriminator"
tokenizer = ElectraTokenizer.from_pretrained(model_name)
model = ElectraForSequenceClassification.from_pretrained(model_name)
```

```python
# total test
def_yes = 0
def_no = 0
error_sentence = [] # 오류난 문장 목록
quiz_sentence = [] # 문제 후보 목록


for text in check_lst:
    # 문장을 토큰으로 분리
    tokens = tokenizer(text, return_tensors="pt")

    # 모델을 사용하여 문법성 및 문맥 판단
    with torch.no_grad():
        outputs = model(**tokens)
        logits = outputs.logits

    # 결과 해석
    predicted_class = logits.argmax().item()
    if predicted_class == 1:
        def_yes += 1
        print("---------------------------------------------------------------")
        print(text)
        print("이 문장은 문법적으로 올바릅니다.")
        quiz_sentence.append(text)
    else:
        print("---------------------------------------------------------------")
        print(text)
        print("이 문장은 문법 오류 또는 문맥 관련 문제가 있을 수 있습니다.")
        error_sentence.append(text)
        def_no += 1
```

```python
print("문장은 문법적으로 올바릅니다.", def_yes)
print("문장에 문법 오류가 있을 수 있습니다. 맞춤법 검사를 수행하세요.", def_no)

print("---------------------------------------------------------------------------------")
print("오류없는 문장 : 109, 오류없다고 평가된 문장 : ",def_yes)
print("오류 문장 : 0, 오류있다고 평가된 문장 : ",def_no)
ghkr = def_yes / 109 * 100
print(ghkr)
```
