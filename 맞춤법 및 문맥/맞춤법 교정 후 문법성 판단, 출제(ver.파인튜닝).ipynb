```python
from transformers import ElectraTokenizer, ElectraForSequenceClassification
import torch

# 사전 훈련된 토크나이저 로드
tokenizer = ElectraTokenizer.from_pretrained("monologg/koelectra-base-v3-discriminator")

# 사전 훈련된 모델 구성 로드
model_config = ElectraConfig.from_pretrained("monologg/koelectra-base-v3-discriminator")

# 모델 가중치가 저장된 파일 경로
model_weights_path = "C:\\Users\\admin\\Desktop\\best.pt"

# 사전 훈련된 모델 구성에 기반하여 새 모델 인스턴스 생성
model = ElectraForSequenceClassification(config=model_config)

# 모델 가중치 로드
model.load_state_dict(torch.load(model_weights_path, map_location=torch.device('cpu')))
```

```python
# 모델을 평가 모드로 설정합니다.
model.eval()

# 결과를 저장할 리스트
results = []
def_yes = 0
def_no = 0
error_sentence = [] # 오류난 문장 목록
quiz_sentence = [] # 문제 후보 목록

# 리스트의 모든 문장을 처리합니다.
for sentence in check_lst:
    # 문장을 토큰화하고 PyTorch 텐서로 변환합니다.
    inputs = tokenizer.encode_plus(
        sentence,
        add_special_tokens=True,
        max_length=128,
        padding='max_length',
        return_attention_mask=True,
        return_tensors='pt'
    )

    # 텐서를 CPU로 이동합니다.
    input_ids = inputs['input_ids'].to('cpu')
    attention_mask = inputs['attention_mask'].to('cpu')

    # 그래디언트 계산을 방지하기 위해 평가 모드로 설정합니다.
    with torch.no_grad():
        # 모델로부터 예측을 수행합니다.
        outputs = model(input_ids, token_type_ids=None, attention_mask=attention_mask)

    # 로짓을 추출하고 소프트맥스를 적용하여 예측 확률을 얻습니다.
    logits = outputs.logits
    predictions = torch.softmax(logits, dim=-1)

    # 예측된 클래스를 추출합니다 (가장 높은 확률을 가진 클래스).
    predicted_class = torch.argmax(predictions).item()

    # 결과를 해석하여 리스트에 추가합니다.
    if predicted_class == 0:
        def_yes += 1
        results.append((sentence, "문법적으로 올바름"))
        quiz_sentence.append(sentence)
    else:
        def_no += 1
        results.append((sentence, "문법적 오류 있음"))
        error_sentence.append(sentence)
        
# 결과 출력
for sentence, result in results:
    print(f"'{sentence}' -> {result}")
    print("---------------------------------------------------------------------------")
print(f"문법적으로 올바른 문장 수: {def_yes}")
print(f"문법적 오류가 있는 문장 수: {def_no}")
ghkr = def_yes / 222 * 100
print(ghkr)
```
